## Numpy array and basics
learn about numpy

```python
import numpy as np
```

### creating array from list

```python
arr_1d = np.array([1, 2, 3])
print("1D array: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array: ", arr_2d)
```

### List vs Numpy array

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)

np_array = np.array([1, 2, 3]) #element wise multiplication
print("Python array multiplication ", np_array * 2)
```

```python
import time
start = time.time()
py_list = [i*2 for i in range(10000000)]
end = time.time()
print("\n List operation time: ", end - start)
```

```python
start_np = time.time()
np_array = np.arange(100000000) * 2
print(np_array)
end_np = time.time()
print("\n Numpy operation time: ", end_np - start_np)
```

### creating array from scratch

```python
zeros = np.zeros((3, 4))
print("zeros array: \n",zeros)

ones = np.ones((3, 4))
print("ones array: \n",ones)

full = np.full((3, 4), 7)
print("full array: \n",full)
```

```python
random = np.random.random((2, 3))
print(random)
```

```python
sequence = np.arange(0, 10)
print(sequence)
```

### Vector, Matrix, Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)
```

```python
matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])
print("Matrix: ", matrix)
```

```python
tensor = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])
print("Tensor: ", tensor)
```

### Array properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6.2]])
print(arr)
print("Shape ", arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("Datatype ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print("Original array: ", arr)
print("Shape: ", arr.shape)

reshaped = arr.reshape([3, 4])
print("Reshaped: ", reshaped)
print("Shape: ", reshaped.shape)

flatten = arr.flatten()
print("flatten: ", flatten)

#ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("Raveled: ", raveled)

# Transpose
transpose = reshaped.transpose()
print("Transpose: ", transpose)
print("Shape: ", transpose.shape)
```
